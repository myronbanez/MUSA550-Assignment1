```python
import pandas as pd

%matplotlib inline
```

# Loading ZHVI data

### Loading in ZHVI data


```python
zhvi = pd.read_csv("./data/Zip_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv")

zhvi
```

### Selecting rows that only display Philadelphia data


```python
studyarea = [
    "Philadelphia County"
]

zhvi.query("CountyName in @studyarea")
```

# Data manipulation and set up

### Creating the the Philly ZHVI dataframe


```python
philly =  zhvi['CountyName'].isin(studyarea)

zhvi_philly = zhvi.loc[philly]

zhvi_philly
```

### Removing unwanted columns to create a cleaner datafrae


```python
trash = ['RegionID', 'SizeRank', "RegionType", "StateName", "State", "Metro", "CountyName", "City"]

zhvi_philly_clean = zhvi_philly.drop(trash, axis=1)

zhvi_philly_clean
```

### Converting dataframe from wide to long


```python
zhvi_philly_df = pd.melt(
    zhvi_philly_clean, 
    id_vars=["RegionName"],
    value_vars=zhvi_philly_clean.columns[1:],
    value_name="ZHVI", 
    var_name="Date"
)

zhvi_philly_df
```

### Adjusting the date column


```python
zhvi_philly_df["Date"] = pd.to_datetime(zhvi_philly_df["Date"])

zhvi_philly_df['Month'] = zhvi_philly_df['Date'].dt.month

zhvi_philly_df['Year'] = zhvi_philly_df['Date'].dt.year 
```

### Calculating annual ZHVI per year and zip code


```python
annual_zhvi_philly = zhvi_philly_df.groupby(['RegionName', 'Year'])['ZHVI'].mean()

annual_zhvi_philly
```

### Indexing the dataframe


```python
annual_zhvi_philly = annual_zhvi_philly.reset_index()

annual_zhvi_philly
```

# Final Dataframe

### After gathering the dataframe, the next step is to group by year and ZHVI then using idx.max to return the zip code with the highest ZHVI by year.

```python
annual_zhvi_philly_1 = annual_zhvi_philly.loc[annual_zhvi_philly.groupby(['Year'])['ZHVI'].idxmax()]
annual_zhvi_philly_1
```
